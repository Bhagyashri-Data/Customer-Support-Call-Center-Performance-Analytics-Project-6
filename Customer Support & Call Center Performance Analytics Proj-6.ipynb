```python
#Task 12: Write an SQL query to find the Total Number of Calls for each Satisfaction_Status. Your output should show two columns: Satisfaction_Status (using the same CASE WHEN logic from Task 11) and Total_Calls
```

```python
query12 = """
SELECT CASE 
            WHEN CSAT_Rating >= 4 THEN 'Satisfied' 
            ELSE 'Dissatisfied' 
       END AS 'Satisfaction_Status',
       COUNT(Call_ID) AS 'Total_Calls'
FROM call_logs
GROUP BY CASE 
              WHEN CSAT_Rating >= 4 THEN 'Satisfied' 
              ELSE 'Dissatisfied' 
         END
"""
```

```python
result12 = pd.read_sql(query12,cust_conn)
result12
```
